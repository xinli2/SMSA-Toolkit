```python
import pandas as pd
import numpy as np
import os
from textblob import TextBlob
from vaderSentiment.vaderSentiment import  SentimentIntensityAnalyzer
```

Part I

I use two packages to analysis the sentiment of the comment from the tweeter data.

the method is from the vader-Sentiment package which consists of LSTM, BERT and other models.

```python
sid_obj = SentimentIntensityAnalyzer()
data_all=pd.DataFrame()
k=0
for file_name in os.listdir('./tweeter_data'):
    try:
        file_location='./tweeter_data'+'/'+file_name
        data=pd.read_csv(file_location)
        data=data['text']
        data.index=range(len(data))
        total_sentiment=0
        for j in range(len(data)):
            sentence=data[j]
            sentiment_dict = sid_obj.polarity_scores(sentence)
            total_sentiment=total_sentiment+sentiment_dict['compound']
        if total_sentiment>0:
            data_all.loc[k,'sentiment']='positive'
        elif total_sentiment<0:
            data_all.loc[k,'sentiment']='negative'
        else:
            data_all.loc[k,'sentiment']='neutral'
        k=k+1
        print(str(k)+ ' is ok')
    except Exception as e:
        pass
```

Part II

we scan the data from the stackoverflow data and find the advice of the product.

```python
sid_obj = SentimentIntensityAnalyzer()
df_sof= pd.read_csv('./data/answers.csv')
df_sof=df_sof['body']
df_sof=pd.DataFrame(df_sof,index=range(len(df_sof)))
for i in range(len(df_sof)):
    df_sof.loc[i,'score']=sid_obj.polarity_scores(df_sof['body'][i])['compound']
df_sof=df_sof.sort_values('score')
```
